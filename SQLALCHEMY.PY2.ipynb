```python
estaciones_df=pd.read_csv(estaciones)
estaciones_df.head()
```

```python
estaciones_df=estaciones_df[["stop_station","stop_lat","stop_lon"]]
estaciones_df.head()
```

```python
last_estaciones_df=estaciones_df.rename(columns={"stop_station":"Station","stop_lat":"Latitude","stop_lon":"Longitude"})
last_estaciones_df.head()
```

```python
descending_estaciones_df=last_estaciones_df.sort_values("Station",ascending=True)
descending_estaciones_df.head()
```

```python
no_index_estaciones=descending_estaciones_df.reset_index(drop=True)
no_index_estaciones.head()
```

```python
last_estaciones=no_index_estaciones.drop_duplicates(subset=['Station'])
last_estaciones.reset_index(drop=True)
last_estaciones.head()
```
